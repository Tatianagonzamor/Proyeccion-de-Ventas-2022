```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv('/content/drive/MyDrive/sales_predictions.csv')
df.head()
```

```python
df.info()
```

```python
Filtro_datos=~df['Item_Weight'].isna() #Filtro de todo lo que o es nulo
Filtro_datos
df2=df.loc[Filtro_datos] # hace el filtro en el dataframe completo
df2.info()
```

```python
df3=df2[["Item_Weight","Item_Identifier"]].groupby(["Item_Identifier"]).mean() #Sacando elpromedio de PESO Según el Identificador
df3.loc['FDC14'] #[] para acceder a una columna o indice / iloc= con indice  númerico aunque no lo tenga/ loc= con el nombre del registro
#df3.index
```

```python
d4=df['Item_Weight'].fillna(df[["Item_Weight","Item_Identifier"]].groupby(["Item_Identifier"])["Item_Weight"].transform('mean'))
#df['Weight'].fillna(df.groupby("Class")['Weight'].transform("mean"))
d4
```

```python
df_final=df.copy() #para copiar el df original, no se cambia el original
df_final['Item_Weight']=d4
df_final.info()
```

```python
Filtro_datos2=df_final['Item_Weight'].isna() #Filtro de todo lo que es nulo
Filtro_datos2
df4=df_final.loc[Filtro_datos2] # hace el filtro en el dataframe completo
df4
```
